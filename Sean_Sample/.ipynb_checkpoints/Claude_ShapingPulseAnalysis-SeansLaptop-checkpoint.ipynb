## 1. Importing Data

```python
import numpy as np
import h5py  # h5py is a package for interface with HDF5 data files
import sys
sys.path.append(r'C:\Users\smc241\OneDrive - University of Cambridge\StranksLab-CEB-107-112-WDO-2\Experiments\diamond\Beamtime Dec 2025\Pulse data')

from Functions import Shaping_amp
from Functions import DataSmoothingShaping
import matplotlib.pyplot as plt
```

```python
# Import all HDF5 files from 0001 to 0041
hdf5_files = [f'Sean_CG100_ST10us_600V_electrons_1DY_50um_{i:04d}.hdf5' for i in range(1, 42)]

PreAmpPulses_all     = []
ShapingAmpPulses_all = []
TriggerTimings_all   = []

for filepath in hdf5_files:
    with h5py.File(filepath, 'r') as f:
        # Read metadata from the first file only (same for all files)
        if not PreAmpPulses_all:  # first iteration
            Timebase  = f['/metadata'].attrs['timebase']
            ChARange  = f['/metadata'].attrs['channel_a_range']
            ChDRange  = f['/metadata'].attrs['channel_d_range']
            for attr_name in f['/metadata'].attrs:
                print(attr_name, f['/metadata'].attrs[attr_name])

        PreAmpPulses_all.append(np.array(f['adc_counts_0']))
        ShapingAmpPulses_all.append(np.array(f['adc_counts_3']))
        TriggerTimings_all.append(np.array(f['trigger_timings']))

# Concatenate across all files
PreAmpPulses     = np.concatenate(PreAmpPulses_all,     axis=0)
ShapingAmpPulses = np.concatenate(ShapingAmpPulses_all, axis=0)
TriggerTimings   = np.concatenate(TriggerTimings_all,   axis=0)

print(f'\nTotal captures loaded: {len(PreAmpPulses)}')
```

```python
# Calculate time axis from timebase metadata
if Timebase > 4: t = (Timebase - 3) * 16
else:            t = (2 ** Timebase)

N_Datapoints = len(PreAmpPulses[0])
Time_us = np.arange(0, t * N_Datapoints, t) / 1e3

N_Captures = len(PreAmpPulses)
print(f'N_Captures: {N_Captures}')
print(f'N_Datapoints per pulse: {N_Datapoints}')
```

```python
# Convert both channels from ADC counts to mV
PreAmpPulses_mV     = PreAmpPulses     * ChARange / 32767
ShapingAmpPulses_mV = ShapingAmpPulses * ChDRange / 32767
```

## 2. Initial Data Visualisation

```python
# Plot 4 evenly spaced example preamp pulses
N_ExamplePulses = 4
ExamplePulseIndices = np.linspace(0, N_Captures - 1, N_ExamplePulses, dtype=int)

fig, axes = plt.subplots(1, 4, figsize=(8, 2.5), tight_layout=True)
for i in range(len(axes)):
    axes[i].set_xlabel(r'Time ($\mu$s)', fontweight='bold')
    axes[i].set_ylabel('Signal (mV)', fontweight='bold')
    axes[i].set_xlim(0, Time_us[-1])
    axes[i].plot(Time_us, PreAmpPulses_mV[ExamplePulseIndices[i]], linewidth=0.1)
plt.show()
```

```python
# Plot 4 evenly spaced example shaping amp pulses
fig, axes = plt.subplots(1, 4, figsize=(8, 2.5), tight_layout=True)
for i in range(len(axes)):
    axes[i].set_xlabel(r'Time ($\mu$s)', fontweight='bold')
    axes[i].set_ylabel('Signal (mV)', fontweight='bold')
    axes[i].set_xlim(0, Time_us[-1])
    axes[i].plot(Time_us, ShapingAmpPulses_mV[ExamplePulseIndices[i]], linewidth=0.1)
plt.show()
```

## 3. Shaping Amplifier Processing

```python
import random

# Smooth all shaping amp pulses across all files
SmoothedShapingPulses = DataSmoothingShaping.WhittakerSmooth(ShapingAmpPulses_mV, lmbd=1e4, order=1)

# Plot 4 evenly spaced examples of raw vs smoothed
fig, axes = plt.subplots(1, 4, figsize=(10, 2.5), tight_layout=True)
for i in range(4):
    axes[i].plot(Time_us, SmoothedShapingPulses[ExamplePulseIndices[i]], color='orange', linewidth=2.5, label='Smoothed pulse')
    axes[i].plot(Time_us, ShapingAmpPulses_mV[ExamplePulseIndices[i]],   color='blue',   linewidth=0.5, label='Raw pulse')
    axes[i].set_xlabel(r'Time ($\mu$s)')
    axes[i].set_ylabel('Signal (mV)')
    axes[i].legend(fontsize=7, loc='best')
plt.show()

print(f'ShapingAmpPulses_mV shape: {ShapingAmpPulses_mV.shape}')
```

```python
# Plot 4 randomly sampled pulses with accepted peak markers
PeakIndex = np.array([np.argmax(pulse) for pulse in ShapingAmpPulses_mV])
print(f'Found {len(PeakIndex)} peaks')

sample_indices = random.sample(range(len(PeakIndex)), 4)
for idx in sample_indices:
    Shaping_amp.plot_accepted_pulses(
        data=ShapingAmpPulses_mV[idx],
        smoothed_data=SmoothedShapingPulses[idx],
        accepted_peaks=[int(PeakIndex[idx])],
        threshold=3,
        window_size=200,
        time_axis=Time_us,
    )
```

```python
# Extract pulse heights using find_pulse_height for all pulses across all files
all_pulse_heights = []
accepted_count    = 0
rejected_count    = 0

MIN_PULSE_HEIGHT = 0.1  # mV

for i, single_peak in enumerate(PeakIndex):
    ph, peak_start_idx, base, top = Shaping_amp.find_pulse_height(
        raw_data=np.array(ShapingAmpPulses_mV[i]).flatten(),
        smoothed_data=np.array(SmoothedShapingPulses[i]).flatten(),
        peak_index=int(single_peak),
        n_sigma=3,
        min_baseline_samples=200,
        baseline_window=200,
    )
    if ph is None or np.isnan(ph):
        rejected_count += 1
    elif abs(ph) < MIN_PULSE_HEIGHT:
        rejected_count += 1
    else:
        all_pulse_heights.append(ph)
        accepted_count += 1

print(f'Accepted pulses: {accepted_count}')
print(f'Rejected pulses: {rejected_count}')
print(f'Total pulses:    {accepted_count + rejected_count}')
print(f'Successfully extracted {len(all_pulse_heights)} pulse heights!')
```

```python
# Plot pulse height histogram
fig, ax = plt.subplots(figsize=(8, 4), tight_layout=True)

ax.hist(
    np.abs(all_pulse_heights),
    bins=np.linspace(0, 8, 200),
    color='steelblue',
    edgecolor='none',
)

ax.set_xlabel('Pulse Height (mV)', fontweight='bold')
ax.set_ylabel('Frequency', fontweight='bold')
ax.set_title('Shaping Amplifier Pulse Height Spectrum')
ax.set_xlim(0, 8)
ax.grid(True, alpha=0.3)
plt.show()
```

```python
from scipy.optimize import curve_fit

# Build histogram
bin_edges   = np.linspace(0, 8, 180)
heights     = np.abs(all_pulse_heights)
counts, edges = np.histogram(heights, bins=bin_edges)
bin_centres = (edges[:-1] + edges[1:]) / 2

# Find the peak bin
peak_bin_idx = np.argmax(counts)
peak_x       = bin_centres[peak_bin_idx]
peak_count   = counts[peak_bin_idx]
half_max     = peak_count / 2

# Fitting region: right side + left side down to half-maximum
left_mask  = (bin_centres < peak_x) & (counts >= half_max)
right_mask = bin_centres >= peak_x
fit_mask   = left_mask | right_mask

x_fit_data = bin_centres[fit_mask]
y_fit_data = counts[fit_mask] / peak_count  # normalise to 1 at peak

def gaussian(x, mean, sigma):
    return np.exp(-0.5 * ((x - mean) / sigma) ** 2)

p0 = [peak_x, 0.5]

try:
    popt, _ = curve_fit(gaussian, x_fit_data, y_fit_data, p0=p0, maxfev=10000)
    mu, sigma = popt
    sigma = abs(sigma)

    peak_min = mu - 2.5 * sigma
    peak_max = mu + 2.5 * sigma

    mask        = (heights >= peak_min) & (heights <= peak_max)
    peak_values = heights[mask]
    centroid    = np.sum(peak_values) / len(peak_values)

    print(f'Fitted Gaussian: \u03bc = {mu:.4f} mV, \u03c3 = {sigma:.4f} mV')
    print(f'Peak window     : {peak_min:.4f} \u2013 {peak_max:.4f} mV  (\u03bc \u00b1 2.5\u03c3)')
    print(f'Counts in window: {len(peak_values)}')
    print(f'Peak centroid   : {centroid:.4f} mV')

    # Plot
    x_curve = np.linspace(0, 8, 500)
    y_curve = peak_count * gaussian(x_curve, mu, sigma)

    fig, ax = plt.subplots(figsize=(8, 4), tight_layout=True)
    ax.hist(heights, bins=bin_edges, color='steelblue', edgecolor='none', label='Data')
    ax.bar(bin_centres[fit_mask], counts[fit_mask],
           width=bin_edges[1] - bin_edges[0], color='cornflowerblue',
           edgecolor='none', label='Bins used for fit')
    ax.plot(x_curve, y_curve, color='crimson', linewidth=2,
            label=f'Gaussian fit (\u03bc={mu:.3f}, \u03c3={sigma:.3f}) mV')
    ax.axhline(half_max, color='grey', linewidth=1, linestyle=':',
               label=f'Half-maximum ({half_max:.0f} counts)')
    ax.axvline(centroid, color='orange', linewidth=1.5, linestyle='--',
               label=f'Centroid = {centroid:.4f} mV')
    ax.axvspan(peak_min, peak_max, alpha=0.1, color='crimson',
               label=f'Peak window (\u03bc \u00b1 2.5\u03c3)')
    ax.set_xlabel('Pulse Height (mV)', fontweight='bold')
    ax.set_ylabel('Frequency', fontweight='bold')
    ax.set_title('Shaping Amplifier Pulse Height Spectrum')
    ax.set_xlim(0, 8)
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.show()

except RuntimeError:
    print('Gaussian fit failed \u2014 try adjusting the initial sigma guess in p0')
```
