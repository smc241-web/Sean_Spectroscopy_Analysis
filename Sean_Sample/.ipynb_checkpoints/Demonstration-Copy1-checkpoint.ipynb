```python
#This code extracts three datasets from the HDF5 file and converts them into NumPy arrays for analysis.
#The other three groups in the hdf5 file contain: data from the preamp, the shaping amp, and the time intervals between which each trigger was recorded
PreAmpPulses     = np.array(TestFile['adc_counts_0'])
ShapingAmpPulses = np.array(TestFile['adc_counts_3'])
TriggerTimings   = np.array(TestFile['trigger_timings'])
```

```python
#The time range for each pulse can be calculated from the timebase property of the metadata
Timebase = TestFile['/metadata'].attrs['timebase']
if Timebase > 4: t = (Timebase-3)*16
else:            t = (2**Timebase)

N_Datapoints = len(PreAmpPulses[0]) 
Time_us = np.arange(0, t*N_Datapoints, t)/1e3
```

```python
#The length of the preamp data array gives the number of triggers recorded for the dataset

#Recall that `PreAmpPulses` is a 2D NumPy array, where:
# - Each **row** is one recorded pulse (a single trigger event)
# - Each **column** is one ADC sample within that pulse

N_Captures = len(PreAmpPulses)
print(N_Captures)
```

```python
#The preamp and shaping amp data can be converted from adc to mV using the range property of the metadata
ChARange = TestFile['/metadata'].attrs['channel_a_range']
ChDRange = TestFile['/metadata'].attrs['channel_d_range']

#These fetch the voltage range settings for two channels from the metadata, channel A and channel D
#The range represents the maximum voltage the ADC was set to record

###pulse = pulse*range/32767

#This shows how to convert a pulse from ADC counts → millivolts. The key number here is 32767, which is 2^15 - 1 — the maximum value of a 16-bit signed integer:
#Dividing by 32767 normalises the pulse to a value between -1 and 1
#Multiplying by range then scales it to the actual voltage in mV
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
#This is a pulse quality control function that filters out bad pulses from the dataset, keeping only clean, well-formed single pulses for analysis.

def PeakAcceptance(RawData, SmoothedData, PkProminence=0.6, PkHeight=0.1, Plots=False):

#RawData — the background corrected pulses
#SmoothedData — the Whittaker-smoothed pulses
#Plots=False — optionally plot rejected pulses for inspection
    
    AcceptedPulseList = []
    PeakIndexList = []
    PileupCounter = 0
    BadTriggerCounter = 0
    UnstableBackgroundCounter = 0
    PulseCounter = 0
    for RawPulse, SmoothedPulse in zip(RawData, SmoothedData):
        
        #Iterates over both arrays simultaneously using zip(), pairing each raw pulse with its corresponding smoothed version.
        
        derivative = np.gradient(SmoothedPulse)
        PeakIndex, PeakProperties = find_peaks(derivative, prominence=PkProminence, height=PkHeight)
        N_Peaks = len(PeakIndex)
        # Check for data sets without a pulse
        if N_Peaks == 0:        
            BadTriggerCounter += 1
            if Plots and BadTriggerCounter<5:      
                fig, (ax1, ax2) = plt.subplots(1,2, figsize=(6, 2))
                ax1.plot(RawPulse)
                ax1.plot(SmoothedPulse)
                ax1.set_title("No Pulse")
                ax2.plot(derivative)
                plt.show()    
        # Check for pile up (data sets with multiple peaks)
        elif N_Peaks >1:
            PileupCounter +=1
            if Plots and PileupCounter<5:
                print("Multiple Pulses")
                fig, (ax1, ax2) = plt.subplots(1,2, figsize=(6, 2))
                ax1.plot(RawPulse)
                ax1.plot(SmoothedPulse)
                ax1.scatter(PeakIndex, SmoothedPulse[PeakIndex], marker="x", color="red", zorder=10)
                ax1.set_title("No Pulse")
                ax2.plot(derivative)
                plt.show() 
                print(PeakProperties)
        # Check for data sets where pre-pulse signal hasn't returned to baseline
        elif N_Peaks == 1:
            if np.argmin(SmoothedPulse)>np.argmax(SmoothedPulse):
                UnstableBackgroundCounter +=1
                if Plots and UnstableBackgroundCounter<5:
                    fig, ax = plt.subplots(1,1, figsize=(3, 2))
                    ax.plot(RawPulse)
                    ax.plot(SmoothedPulse)
                    ax.set_title("Baseline Instability")
                    plt.show()
            else:
                AcceptedPulseList.append(RawPulse)
                PeakIndexList.append(PeakIndex)
                PulseCounter +=1
                if Plots and PulseCounter<5:
                    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(6, 2))
                    ax1.plot(RawPulse)
                    ax1.plot(SmoothedPulse)
                    ax1.scatter(PeakIndex, SmoothedPulse[PeakIndex], marker="x", color="red", zorder=10)
                    ax1.set_title("Accepted Pulse")
                    ax2.plot(derivative)
                    plt.show()    
                    print(PeakProperties)
          
    Counters = [BadTriggerCounter, PileupCounter, UnstableBackgroundCounter, PulseCounter]
    CounterNames = ["Data Sets with no Pulse:", "Pulses Rejected for Pileup:", "Pulses Rejected for Baseline Instability:", "Accepted Pulses:"]
    for name, count in zip(CounterNames, Counters): print(name, count)
    if BadTriggerCounter + PileupCounter + UnstableBackgroundCounter + PulseCounter != len(RawData):
        print("Warning: Missing Pulses")
    return AcceptedPulseList, PeakIndexList
```

```python
#Locate pulse boundaries where derivative returns to baseline
def PulseBoundaries(derivative, PeakIndex, window=10, threshold=0.02):

#derivative of the smoothed pulse calculated
#window=10 — number of samples to average over when checking the baseline
#threshold=0.02 — how close to zero the derivative must be to count as baseline
    
    # ---- Find pulse start (scan backwards from peak) ----
    # Start at the peak index and move left (toward index 0)
    # Stop once we reach `window` to avoid invalid slicing.
    for i in range(PeakIndex[0], window, -1):
        # Compute mean derivative in the window immediately to the left
        # of the current index. This smooths noise fluctuations.
        local_mean = np.mean(derivative[i-window:i])
               
        # If derivative has returned close to baseline level,
        # we assume this marks the start of the pulse.
        if local_mean < threshold:
            # Define pulse start as the midpoint of that window.
            # This centers the boundary estimate
            PulseStart = i - window//2
            break
    # If the loop completes without finding a boundary,
    # the `for-else` construct triggers this return.
    else: return None, None

    # ---- Find pulse end (scan forwards from peak) ----
    # Move right from the peak toward the end of the array.
    # Stop early enough so slicing does not exceed array bounds.
    for i in range(PeakIndex[0], len(derivative)-window):
        # Compute mean derivative in the window immediately to the right.
        local_mean = np.mean(derivative[i:i+window])
        # If derivative returns below threshold, pulse tail has ended.
        if local_mean < threshold:
            # Again, choose midpoint of window as boundary.
            PulseEnd = i + window//2
            break
    else: return None, None
    
    # Return detected boundaries
    return PulseStart, PulseEnd
```

```python
def PulseHeight(PulseData, PulseStart, PulseEnd):
    Baseline = PulseData[:PulseStart]
    Top      = PulseData[PulseEnd:]

    PH = np.mean(Top) - np.mean(Baseline)

    PH_Err = np.sqrt((np.std(Top)/np.sqrt(len(Top)))**2 + 
                      (np.std(Baseline)/np.sqrt(len(Baseline)))**2)

    return PH, PH_Err, Baseline, Top
```

```python
def RiseTime(PulseData, PeakIndex, PH, Time, window=20):

    h10 = 0.1 * PH
    h90 = 0.9 * PH

    t10 = None
    t90 = None

    # Find 10% crossing
    for i in range(PeakIndex, window, -1):
        if np.mean(PulseData[i-window:i]) < h10:
            t10 = Time[i]
            break

    # Find 90% crossing
    for i in range(PeakIndex, len(PulseData)-window):
        if np.mean(PulseData[i:i+window]) > h90:
            t90 = Time[i]
            break

    if t10 is None or t90 is None or t90 <= t10: return None, None, None
    RT = (t90 - t10) / 0.8

    return RT, t10, t90
```

```python
def SinglePulseCharacterisation(RawPulse, SmoothedPulse, PeakIndex, Time, BoundaryWindow=10, BoundaryThreshold=0.02, Plots=False, RejectCounter=0, PlotCounter=0):
    derivative = np.gradient(SmoothedPulse)

    PulseStart, PulseEnd = PulseBoundaries(derivative, PeakIndex, window=BoundaryWindow, threshold=BoundaryThreshold)
    if PulseStart is None: 
       if Plots and RejectCounter < 5:
            RejectCounter += 1
            print("Pulse boundaries not found")
            fig, (ax1, ax2) = plt.subplots(1,2, figsize=(8,3))
            ax1.plot(Time, RawPulse)
            ax1.plot(Time, SmoothedPulse)
            ax2.plot(Time, derivative)
            plt.show()
       return None
    
    PH, PH_Err, Baseline, Top = PulseHeight(SmoothedPulse, PulseStart, PulseEnd)
    if PH <= 0: return None
    
    CorrectedPulse = SmoothedPulse - Baseline

    RT, t10, t90 = RiseTime(CorrectedPulse, PeakIndex, PH, Time)
    if RT is None:         
       if Plots and RejectCounter < 5:
            RejectCounter += 1
            print("Rise time extraction failed")
            fig, ax = plt.subplots(figsize=(6,3))
            ax1.plot(Time, CorrectedPulse)
            plt.show()
       return None
    
    if Plots and PlotCounter < 5:
        fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,3))

        ax1.plot(Time, RawPulse, alpha=0.5)
        ax1.plot(Time, SmoothedPulse)
        ax1.scatter(Time[PeakIndex], SmoothedPulse[PeakIndex], color="red", zorder=10)
        ax1.axhline(Baseline, linestyle="--")
        ax1.axhline(Top, linestyle="--")
        ax1.axvline(Time[PulseStart], color="grey", linestyle=":")
        ax1.axvline(Time[PulseEnd], color="grey", linestyle=":")

        ax1.set_title(f"Height = {PH:.2f}")

        # Derivative
        ax2.plot(Time, derivative)
        ax2.scatter(Time[PeakIndex], derivative[PeakIndex], color="red", zorder=10)

        # Rise time markers
        ax1.axhline(0.1*PH, linestyle=":")
        ax1.axhline(0.9*PH, linestyle=":")
        ax1.axvline(t10, color="green")
        ax1.axvline(t90, color="green")

        ax2.set_title(f"Rise = {RT:.2f} µs")

        plt.tight_layout()
        plt.show()
    
    return PH, RT, PH_Err, RejectCounter
```

```python
#This function is the top level loop that runs SinglePulseCharacterisation on every pulse in the dataset, collecting the results into arrays for analysis.

def AlphaPulsePeakCharacterisation(ValidPulsesRaw, ValidPulsesSmoothed, peaks, Time_microsec, Plots=False):

    RiseTimes = []
    PulseHeights = []
    PulseHeightErrors = []
    PlotCounter = 0
    RejectCounter = 0

    for raw, smooth, pk in zip(ValidPulsesRaw, ValidPulsesSmoothed, peaks):

        result = SinglePulseCharacterisation(raw, smooth, pk, Time_microsec, Plots=Plots, RejectCounter=RejectCounter, PlotCounter=PlotCounter)

        if result is None:
            RejectCounter += 1
            continue

        rise, height, err = result

        RiseTimes.append(rise)
        PulseHeights.append(height)
        PulseHeightErrors.append(err)

    return (np.array(RiseTimes), np.array(PulseHeights), RejectCounter, np.array(PulseHeightErrors))
```
