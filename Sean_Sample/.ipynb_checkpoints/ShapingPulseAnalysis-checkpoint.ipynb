## 1. Importing Data

```python
import numpy as np
import h5py #h5py is a package for interface with HDF5 data files
```

```python
# Import Test Data File
TestFile =  h5py.File('Sean_CG100_ST10us_600V_electrons_1DY_50um_0041.hdf5','r')

# A hdf5 file is structured like a Python dictionary; the keys tell us what is stored in the file
list(TestFile.keys())
```

```python
#The metadata is stored as attributes within the hdf5 file: here are the names and values of each one
for attr_name in TestFile['/metadata'].attrs: print(attr_name, TestFile['/metadata'].attrs[attr_name])
```

```python
#The other three groups in the hdf5 file contain: data from the preamp, the shaping amp, and the time intervals between which each trigger was recorded
PreAmpPulses     = np.array(TestFile['adc_counts_0'])
ShapingAmpPulses = np.array(TestFile['adc_counts_3'])
TriggerTimings   = np.array(TestFile['trigger_timings'])
```

```python
#The time range for each pulse can be calculated from the timebase property of the metadata
Timebase = TestFile['/metadata'].attrs['timebase']
if Timebase > 4: t = (Timebase-3)*16
else:            t = (2**Timebase)

N_Datapoints = len(PreAmpPulses[0]) 
Time_us = np.arange(0, t*N_Datapoints, t)/1e3
```

```python
#The length of the preamp data array gives the number of triggers recorded for the dataset
N_Captures = len(PreAmpPulses)
print(N_Captures)
```

```python
#The preamp and shaping amp data can be converted from adc to mV using the range property of the metadata
ChARange = TestFile['/metadata'].attrs['channel_a_range']
ChDRange = TestFile['/metadata'].attrs['channel_d_range']

#pulse = pulse*range/32767
```

```python
# Add this after cell 7 — convert both channels to mV
PreAmpPulses_mV      = PreAmpPulses      * ChARange / 32767
ShapingAmpPulses_mV  = ShapingAmpPulses  * ChDRange / 32767
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# --- Build histogram ---
bin_edges = np.linspace(0, 8, 180)
heights = np.abs(all_pulse_heights)
counts, edges = np.histogram(heights, bins=bin_edges)
bin_centres = (edges[:-1] + edges[1:]) / 2

# --- Find the peak bin ---
peak_bin_idx = np.argmax(counts)
peak_x = bin_centres[peak_bin_idx]
peak_count = counts[peak_bin_idx]
half_max = peak_count / 2

# --- Define fitting region ---
# Right side: all bins from peak rightward
# Left side: bins from peak leftward, but only down to half-maximum
left_mask  = (bin_centres < peak_x) & (counts >= half_max)
right_mask = bin_centres >= peak_x

fit_mask = left_mask | right_mask
x_fit_data = bin_centres[fit_mask]
y_fit_data = counts[fit_mask] / peak_count  # normalise to 1 at peak

def gaussian(x, mean, sigma):
    return np.exp(-0.5 * ((x - mean) / sigma) ** 2)

p0 = [peak_x, 0.5]

try:
    popt, _ = curve_fit(gaussian, x_fit_data, y_fit_data, p0=p0, maxfev=10000)
    mu, sigma = popt
    sigma = abs(sigma)

    peak_min = mu - 2.5 * sigma
    peak_max = mu + 2.5 * sigma

    mask = (heights >= peak_min) & (heights <= peak_max)
    peak_values = heights[mask]
    centroid = np.sum(peak_values) / len(peak_values)

    print(f"Fitted Gaussian: μ = {mu:.4f} mV, σ = {sigma:.4f} mV")
    print(f"Peak window     : {peak_min:.4f} – {peak_max:.4f} mV  (μ ± 2.5σ)")
    print(f"Counts in window: {len(peak_values)}")
    print(f"Peak centroid   : {centroid:.4f} mV")

    # --- Plot ---
    x_curve = np.linspace(0, 8, 500)
    y_curve = peak_count * gaussian(x_curve, mu, sigma)

    fig, ax = plt.subplots(figsize=(8, 4), tight_layout=True)
    ax.hist(heights, bins=bin_edges, color="steelblue", edgecolor="none", label="Data")

    # Highlight the bins used for fitting
    ax.bar(bin_centres[fit_mask], counts[fit_mask],
           width=bin_edges[1]-bin_edges[0], color="cornflowerblue",
           edgecolor="none", label="Bins used for fit")

    ax.plot(x_curve, y_curve, color="crimson", linewidth=2,
            label=f"Gaussian fit (μ={mu:.3f}, σ={sigma:.3f}) mV")
    ax.axhline(half_max, color="grey", linewidth=1, linestyle=":",
               label=f"Half-maximum ({half_max:.0f} counts)")
    ax.axvline(centroid, color="orange", linewidth=1.5, linestyle="--",
               label=f"Centroid = {centroid:.4f} mV")
    ax.axvspan(peak_min, peak_max, alpha=0.1, color="crimson",
               label=f"Peak window (μ ± 2.5σ)")
    ax.set_xlabel("Pulse Height (mV)", fontweight="bold")
    ax.set_ylabel("Frequency", fontweight="bold")
    ax.set_title("Shaping Amplifier Pulse Height Spectrum")
    ax.set_xlim(0, 8)
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.show()

except RuntimeError:
    print("Gaussian fit failed — try adjusting the initial sigma guess in p0")
```
